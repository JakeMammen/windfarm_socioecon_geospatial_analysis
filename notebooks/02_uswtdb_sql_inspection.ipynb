```python
# ============================================================
# ROBUST PATH FIX
# ============================================================
import os
from pathlib import Path

try:
    project_root = Path(__file__).parent.parent
except NameError:
    cwd = Path(os.getcwd())
    if cwd.name == "notebooks":
        project_root = cwd.parent
    else:
        project_root = cwd

os.chdir(project_root)
print(f"Working directory set to: {os.getcwd()}")
# ============================================================

import duckdb

con = duckdb.connect()
con.execute("INSTALL spatial; LOAD spatial;")

print("=" * 70)
print("USWTDB Parquet - Column Names and Data Types")
print("=" * 70)

con.sql("""
    DESCRIBE 'data/processed/uswtdb_ia_ok_tx_2026.parquet'
""").show(max_rows=30)

print("=" * 70)
```
